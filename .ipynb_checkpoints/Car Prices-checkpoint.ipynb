```python
# print datatype of any column/feature by name
def print_dtypes(feature_name):
    print(f'{feature_name} has data type {df[feature_name].dtype}')

# converting a whole feature from string to integer price in this case
def conv_to_int(string_value): # $15,500
    convertable_value = string_value[1:].replace(',', '')
    return int(convertable_value)

# traversing through and printing datatype of all columns
for feature in df.columns:
    print_dtypes(feature)

# applying function on dataframe with work like that(conv_to_int is function declared above)
df['price'] = df['price'].apply(conv_to_int)
print(df['price'].dtype)
```

```python
# mean here is important to determine which brand, model or any other feature value of car has probably the most high prices 
def mean_of_seperate_values(column_name, values):
    greater_mean = {'mean': 0,
                   'value': ''}
    for value in values:
        temp_df = df[df[column_name] == value]
        mean = temp_df.price.mean()
        # print(f'price mean for {value} in {column_name} is : ', mean)
        if mean > greater_mean['mean']:
            greater_mean['mean'] = mean
            greater_mean['value'] = value
    print('----------------------Greatest mean is :', greater_mean['mean'], 'for value ', greater_mean['value'])

for feature in df.columns:
    distinct_values = df[feature].unique()
    mean_of_seperate_values(feature, distinct_values)



# this will make output too longer so I only printed greatest mean for each value here
# buggatti's have probably the most high prices.
# you can confirm it by also printing the medain etc
```

```python
# checking nan values from all columns
def find_nan_values(column_name):
    return df[column_name].isna().sum()

for column_name in df.columns:
    print(f'{column_name} has na values : {find_nan_values(column_name)}')


# these columns have nan values so checking thier other unique values
print(df['fuel_type'].unique())
print(df['accident'].unique())
print(df['clean_title'].unique())

# how much null values they have
(df['accident'].isna()).sum()
(df['fuel_type'].isna()).sum()
(df['clean_title'].isna()).sum()
```

#### Based on previous insights
#### -> clean_title feature have nan instead of No values
#### -> accident has both yes or no values and nan where it is unknown
#### -> fuel_type has nan and also '' value representing nan

```python
# fixing clean_title values by replacing nan with no
df.loc[(df['clean_title'].isna()), 'clean_title'] = 'No'
print(df['clean_title'])
```

```python
# now there is a possiblity of having '' instead of nan values
# so checking them
def find_unexpected_values(column_name):
    return ((df[column_name] == '').sum())

for column_name in df.columns:
    print(f'{column_name} has unexpected values : {find_unexpected_values(column_name)}')
```

```python
# printing all modes for columns.
def print_mode(column_name):
    mode = df[column_name].mode()[0]
    mode_count = (df[column_name] == mode).sum()
    return mode, mode_count

for column in df.columns:
    print(print_mode(column))
print(df.columns)
```

```python
# handling nan and unexpected values
def fill_unexpected(column_name):
    # this function returns a tuple/list
    mode_details = print_mode(column_name)
    # mode_details[0] on right is destructing list and assigning its first values i.e. mode to L.H.S.(the nan values)
    df.loc[((df[column_name].isna()) | (df[column_name] == '')), column_name] = mode_details[0]

for column in df.columns:
    if column != 'accident':
        fill_unexpected(column)

# with accident column we will deal later with price we don't need to replace missing values, 
# also not in some other features we dont have any na values so they remains as they are 
```

```python
def log_scaling(feature):
    df[feature] = np.log(df[feature] + 1)
    print(df[feature])

# apply log scaling on price column
log_scaling('price')
print(df['price'])
```

```python
# commented method in this function takes alot of time alternative is using groupby and .map builtin functions
# this function basically find mean of price for every distinct value, and put it(numerical) in place of
# ategorical values   
def categorical_to_numerical(feature_name):
    # method 1
    # df.loc[(df[feature_name] == value), feature_name] = df.loc[(df[feature_name] == value), 'price'].mean()
    
    distinct_values = df[feature_name].unique()
    mean_prices = df.groupby(feature_name)['price'].mean()
    df[feature_name] = df[feature_name].map(mean_prices).fillna(df[feature_name])
    print(mean_prices)
    # print(f'mean value of price for value {value} is {df.loc[(df[feature_name] == value), 'price'].mean()}')



for feature in df.columns:
    if feature != 'price' and feature!= 'accident':
        categorical_to_numerical(feature)

# left accident for now
```

```python
# fixing accident column
# for this we have known values if the car has at least 1 accident then it is considered as bad effect on price,
# so we put 0 here, similarly for no accident we can put it to 1
# for the nan values we don't have any idea so we put 0.5 here

df.loc[(df['accident'] == 'At least 1 accident or damage reported'), 'accident'] = 0
df.loc[(df['accident'] == 'None reported'), 'accident'] = 1
df.loc[(df['accident'].isna()), 'accident'] = 0.5


for feature in df.columns:
    df[feature] = df[feature].apply(float)

df.info()
df.head()
print('for accident nan values are now : ' , df['accident'].isna().sum())
```

```python
# dividing dependent and independent features
y = df['price']
X = df.drop('price', axis=1, inplace=False)


print(y)
print(X)
```

```python
# splitting data into train and test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
# creating a scaler to scale data around 0 and sd 1
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
regressor = LinearRegression()
```

```python
# performing scaling
# scaling will not apply to y because it is to be predicted value
# the main objective of scaling is to shrink the data intervals, while keeping thier relative data same
# therefore it has good affect on accuracy
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
print(X_train_scaled)
print(X_test_scaled)
```

```python
# prepare model
regressor.fit(X_train_scaled, y_train)


```

```python
y_pred = regressor.predict(X_test_scaled)
print(y_pred)
```

```python
# importing error and accuracy finding functions
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import cross_val_score
```

```python
# finding and printing accuracies and errors
r2 = r2_score(y_pred, y_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
cross_val_scores = cross_val_score(regressor, X, y, cv=5, scoring='r2')

print('r squared', r2)
print('mean absolute error', mae)
print('mean squared error', mse)
print('root mean squared error', rmse)
print('cross validation score', cross_val_scores)
```

```python
# plotting y_prediction vs y_test graph
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)  # line for perfect predictions
plt.title('Actual vs. Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.grid()
plt.show()
```
